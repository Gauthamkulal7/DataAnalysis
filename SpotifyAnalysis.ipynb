```python
data.dropna(subset=['valence'], inplace=True)
data.head()
```

```python
#convert the duration form milliseconds to seconds
data["duration"]=data["duration_ms"].apply(lambda x:round(x/1000))
data.drop("duration_ms",inplace=True,axis=1)
data.head()
```

```python
#Converting release_date to index
data.set_index("release_date",inplace=True)
data.index=pd.to_datetime(data.index,infer_datetime_format=True)
data.head()
```

```python
# Count the number of unique genres and their occurrences
genre_counts = data['genre'].value_counts()
top_genre = genre_counts.idxmax()
print("Top Genre:", top_genre)
```

```python
# Group the data by 'genre' and calculate the average 'popularity' for each genre
average_popularity_by_genre = data.groupby('genre')['popularity'].mean()

# Get the top 2 genres by 'popularity'
top_popularity_genres = average_popularity_by_genre.nlargest(2)

# Create a bar graph for the top 2 popularity genres
plt.figure(figsize=(10, 6))
top_popularity_genres.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Top 2 Genres by Popularity')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Group the data by 'genre' and calculate the average 'popularity' for each genre
average_popularity_by_genre = data.groupby('genre')['popularity'].mean()

# Get the bottom 2 genres by 'popularity'
bottom_popularity_genres = average_popularity_by_genre.nsmallest(2)

# Create a bar graph for the bottom 2 popularity genres
plt.figure(figsize=(10, 6))
bottom_popularity_genres.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Bottom 2 Genres by Popularity')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Splitting dependent and independent data
x = data[['loudness','valence']]
y = data[['energy']]

# Splitting dataset into test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.3, random_state = 0)

# built a LinearRegression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)

#checking the score
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print("Score:",score)
```

```python
y_test_pred = regressor.predict(x_test)
ax = sns.scatterplot(x=y_test['energy'], y=y_test_pred.flatten())
sns.lineplot(x=y_test['energy'], y=y_test['energy'], color='tab:orange', ax=ax)
ax.set(xlabel='Predicted energy', ylabel='Predicted energy',
       title='Prediction performance on the test sample')
plt.show()
```
